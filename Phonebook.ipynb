```python
import pymongo

#pymongo.version

from pymongo import MongoClient

client = pymongo.MongoClient("mongodb+srv://phonebook:ph15@cluster0.fpv8y.mongodb.net/myFirstDatabase?retryWrites=true&w=majority")
db = client.test

records=db.phonebook_test_collection
records.count_documents({})
```

```python
# creating contacts
quant=int(input("Enter the number of contacts to be created:"))
mylist=[]
for i in range (0,quant):
  n=input("Please add your name:")
  p=input("Please add your phone number:")
  w=input("Is this your whatsapp number?:")
  if w=="yes":
    print("Your number is added")
    d=p
  elif w=="no":
    d=(input("Please add your whatsapp number:"))
  e=input("Please enter your email address:")
  mylist.append({"Name":n,"Phone Number":p,"Whatsapp":d,"Email":e})
x=records.insert_many(mylist)
print(x.inserted_ids)
```

```python
#searching contact 
myquery={"Name":input()}
mydoc=records.find(myquery,{"_id":0,"Name":1,"Phone Number":1})

for x in mydoc:
  print(x)
```

```python
#searching all contacts for names starting with a certain alphabet
myquery={"Name":{"$regex":"^"+input()}}
for x in records.find(myquery,{"_id":0,"Name":1,"Phone Number":1,"Whatsapp":1,"Email":1}):
  print(x)

```

```python
#deleting a contact
myquery={"Name":input()}
x=records.delete_many(myquery)
print(x.deleted_count,"contacts deleted.")
```

```python
#displaying all contacts in alphabetical order
for x in records.find({},{"_id":0,"Name":1,"Phone Number":1,"Whatsapp":1,"Email":1}).sort("Name",pymongo.ASCENDING):
  print(x)

```
